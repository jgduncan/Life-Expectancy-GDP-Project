```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('all_data.csv')
print(data.head())
print(data.Country.unique())
```

```python
chile = data[data['Country'] == 'Chile']

sns.set_palette('muted')
sns.set_style('white')
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.lineplot(data=data, x='Year', y='GDP', hue='Country', linewidth=3)
ax.set_title('GDP change over time')
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country')
```

```python
sns.set_style('dark')
sns.set_palette('Paired')
plt.figure(figsize=(10, 10))
sns.barplot(data=data, x='Country', y='Life expectancy at birth (years)', ci='sd')
plt.title('Average Life Expectancy between Countries')
plt.figure(figsize=(10, 10))
sns.barplot(data=data, x='Country', y='GDP', ci='sd')
plt.title('Average GDP between Countries')
```

```python
sns.set_palette('Dark2')
sns.set_style('darkgrid')
plt.figure(figsize=(15,10))
plt.suptitle('Life Expecatancy VS GDP Over Time')
plt.subplot(2, 3, 1)
sns.scatterplot(data=data[data.Country == 'Chile'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('Chile')
plt.subplot(2, 3, 2)
sns.scatterplot(data=data[data.Country == 'China'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('China')
plt.subplot(2, 3, 3)
sns.scatterplot(data=data[data.Country == 'Germany'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('Germany')
plt.subplot(2, 3, 4)
sns.scatterplot(data=data[data.Country == 'Mexico'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('Mexico')
plt.subplot(2, 3, 5)
sns.scatterplot(data=data[data.Country == 'United States of America'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('United States of America')
plt.subplot(2, 3, 6)
sns.scatterplot(data=data[data.Country == 'Zimbabwe'], x='Life expectancy at birth (years)', y='GDP', hue='Year')
plt.title('Zimbabwe')
```
